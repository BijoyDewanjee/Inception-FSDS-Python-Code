# EDA - Univariate Analysis

```python
df = pd.read_csv('train.csv')
```

# 1. Categorical Data

## a. Countplot

```python
import seaborn as sns
```

```python
sns.countplot(df, x="Survived")
```

```python
sns.countplot(df, x="Embarked")
```

```python
sns.countplot(df, x="Pclass")
```

```python
df["Survived"].value_counts().plot(kind="pie", autopct="%.2f")
```
